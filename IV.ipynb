```python
from math import pi

ACCURACY = 0.01
ADC_STEPS = 2**18
ADC_VOLTAGE_REFERENCE = 2.048 # V
LSB = ADC_VOLTAGE_REFERENCE/ADC_STEPS # V
DIVISION_RATIO = 2.5

MINIMUM_ADC_VOLTAGE = (LSB/ACCURACY)
MAXIMUM_ADC_VOLTAGE = (ADC_VOLTAGE_REFERENCE - MINIMUM_ADC_VOLTAGE)

RESISTANCE_RANGES = [1e3, 10e3]
f = 100e3
V = 5
TRANSIMPEDANCE_GAIN = 1e3

capacitance_range_order_of_magnitude = [10**(-1*i) for i in range(7, 13)]
inductance_range_order_of_magnitude = [10**(-1*i) for i in range(3, 10)]

print("CAPACITANCE")
for c in capacitance_range_order_of_magnitude:
    for r in RESISTANCE_RANGES:
        Xc = 1/(2*pi*f*c)
        Z = r - 1j*Xc
        I = V/Z
        voltage_signal = abs(-1j*Xc*I)/DIVISION_RATIO
        current_signal = (abs(I)*TRANSIMPEDANCE_GAIN)/DIVISION_RATIO
        current_signal_satisfactory = current_signal < MAXIMUM_ADC_VOLTAGE and current_signal > MINIMUM_ADC_VOLTAGE
        voltage_signal_satisfactory = voltage_signal < MAXIMUM_ADC_VOLTAGE and voltage_signal > MINIMUM_ADC_VOLTAGE

        if(current_signal_satisfactory and voltage_signal_satisfactory):
            print(r, c)

print("INDUCTANCE")
for l in inductance_range_order_of_magnitude:
    for r in RESISTANCE_RANGES:
        Xl = 2*pi*f*l
        Z = r + 1j*Xl
        I = V/Z
        Vout = 1j*Xl*I

        
        voltage_signal = abs(-1j*Xl*I)/DIVISION_RATIO
        current_signal = (abs(I)*TRANSIMPEDANCE_GAIN)/DIVISION_RATIO
        current_signal_satisfactory = current_signal < MAXIMUM_ADC_VOLTAGE and current_signal > MINIMUM_ADC_VOLTAGE
        voltage_signal_satisfactory = voltage_signal < MAXIMUM_ADC_VOLTAGE and voltage_signal > MINIMUM_ADC_VOLTAGE

        if(current_signal_satisfactory and voltage_signal_satisfactory):
            print(r, l)
```
