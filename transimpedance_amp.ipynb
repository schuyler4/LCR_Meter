```python
from stability_analysis import evaluate_TF, plot_TF
from math import pi
```

```python
GBWP = 6e6
OPEN_LOOP_GAIN = 80000
R = 1000
C = 1.3e-9

def Xc(c): return 1/(2*pi*100e3*c)

closed_loop_gain = R/Xc(C)
cutoff_frequency = GBWP/closed_loop_gain

def TF(s): return OPEN_LOOP_GAIN/((1+(s/cutoff_frequency))*(1+s*R*C))
response =  evaluate_TF(TF, 1e6,100e6)
```

```python
plot_TF(response)
```
